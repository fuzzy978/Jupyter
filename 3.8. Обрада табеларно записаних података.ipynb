# Обрада табеларно записаних података

## Задатак 1
Табела садржи податке са именима ученика и закључним оценама из неких предмета на крају првог полугодишта.

1. Податке учитати у низ razred.
2. Податке приказати табеларно.

## Задатак 2
1. Креирај нову табелу која ће као индекс имати колону Ime.
2. Додај нову колону Prosek која ће садржати просечну оцену сваког ученика заокружену на две децимале.
3. Додај нови ред ProsekPr који ће садржати просечну оцену за сваки предмет.

## Задатак 3
1. Исписати имена ученика и њихов појединачни просек.
2. Исписати просечну оцену за сваки предмет појединачно.
3. Исписати све оцене једног од ученика, као и његову просечну оцену.
