# Задатак 1
1. Направити два низа, gradovi и prosecne_temperature. 

2. Нацртати линијски график. На x оси се налазе градови а на y оси просечне температуре.

## Задатак 2
1. Нацртати графикон са кружићима, који представља квадрат бројева 1, 2, 3 и 4. На x оси се налазе бројеви 0-6, а на y оси од 0-20.

## Задатак 3
1. Направити два низа, takmicari и bodovi.

2. Нацртати стубичасти график. На x оси се налазе такмичари а на y оси њихови бодови.
3. Нацртати стубичасти график, ширине 14 и висина 7 инча. Боја стубића нека буде магента.

## Задатак 4

1. Направити два низа, mreze и odgovori.
2. Нацртати секторски (пита) графикон, ширине 5 и висине 5 инча, као наслов написати Korišćenje društvenih mreža.
